This project aims to predict whether a customer will purchase a given product based on his/her age and estimated salary using Logisitic regression. I have also calculated the accuracy of this model using confusion matrix

```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
# Importing the datasets
df = pd.read_csv("Social_Network_Ads.csv")
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(x,y)
```

```python
# Insights of our datasets
df.describe()
```

```python
# Visualizing the dataset
a = df.iloc[:,0].values
b = df.iloc[:,1].values
print(a,b,y)

sn.boxplot(x=y, y=a)
plt.show()


sn.boxplot(x=y,y=b)
plt.show()



```

```python
# spliting the data
from sklearn.model_selection import train_test_split
x1,x2,y1,y2 = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x1 = sc.fit_transform(x1)
x2 = sc.fit_transform(x2)
```

```python
# Applying logisitic reg
from sklearn.linear_model import LogisticRegression

cl = LogisticRegression(random_state =0)   #cl = classifier()
cl.fit(x1,y1)
```

```python
# Predicting the Test cases
yp = cl.predict(x2)
print(y2,yp)
print(np.concatenate((yp.reshape(len(yp),1),y2.reshape(len(y2),1)),1))
```

```python
# MAKING CONFUSING MATRIX TO SEE THE QUALITY OF OUR MODEL
from sklearn.metrics import confusion_matrix, accuracy_score 

# metrics module contains all the necessary tools to check our models accuracy
# R2-SCORE IS USED OF REG.(continous) AND C.M IS USED FOR CLASSIFICATION(discrete)
# True positives (TP): the number of observations that are correctly classified as positive by the model.
# False positives (FP): the number of observations that are incorrectly classified as positive by the model.
# True negatives (TN): the number of observations that are correctly classified as negative by the model.
# False negatives (FN): the number of observations that are incorrectly classified as negative by the model.
cm = confusion_matrix(y2,yp)
print(cm)

# accuracy = (TP+TN)/len(test set)
accuracy_score(y2,yp)

 
```
